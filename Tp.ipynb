```python
#import required libraries
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
Train_v2 = pd.read_csv('Train_v2.csv')
Train_v2
```

```python
# droping the data we don't need
Train_v2 = Train_v2.drop(["uniqueid"], axis = 1)
Train_v2.dropna(inplace=True)
Train_v2
```

```python
y = Train_v2.iloc[:,2].values
Train_v2 = Train_v2.drop(["bank_account"], axis = 1)
X = Train_v2.iloc[:, :].values
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y = label.fit_transform(y)
X[:,0] = label.fit_transform(X[:,0])
X[:,2] = label.fit_transform(X[:,2])
X[:,3] = label.fit_transform(X[:,3])
X[:,6] = label.fit_transform(X[:,6])
X[:,8] = label.fit_transform(X[:,8])
X[:,9] = label.fit_transform(X[:,9])
X[:,10] = label.fit_transform(X[:,10])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print('The score For Logistic Regression: ' , r2_score(y_test , y_pred))
print('The accuracy for LR is ', (y_pred == y_test).mean())
cm = confusion_matrix(y_test,y_pred)
print('Confusion Matrix is :', cm)
```
